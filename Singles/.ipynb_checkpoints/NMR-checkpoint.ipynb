# 实验原理 

###   核磁共振

在没有外磁场时，自旋核的取向是任意的，并且自旋产生的磁场方向也是任意的，自旋量子数相同而磁量子数不同的核，其角动量虽有不同的空间取向，但其能量是相同的，即能级是简并的。如果把核放在外磁场中，由于磁场间的相互作用，在外磁场中核磁矩将有 2I+1 个不同的空间取向，每种取向都对应一个能级状态。

如果在$z$方向均匀磁场$H_{0}$存在的同时，再加上一个方向与之垂直，强度远小于$H_{0}$的射频交变磁场$H_{1}$(设在$x$方向) ，则只要圆频率ω满足下述频率条件：$$\hbar\omega = \Delta E = \omega_{0}$$ 原子核就会吸收射频磁场的能量，在塞曼能级之间跃迁，即共振。

对于共振条件：
- 对于同一种核，磁旋比为定值，$H_{0}$改变，射频频率$\nu$变。
- 不同原子核，磁旋比不同，产生共振的条件不同，需要的磁场强度 $H_{0}$和射频频率$\nu$不同。
- 固定$H_{0}$，改变$\nu$（扫频方法），不同原子核在不同频率处发生共振。也可固定，改变$H_{0}$（扫场方法）。


### 化学位移

核磁共振的共振条件是$\omega=\gamma H_{0}$ 。按照这个公式，同一种原子核，由于旋磁比相同，因而在相同外磁场下只应有一个共振频率。但实际上，由于所处的分子不同或在分子内处于不同的化学基团，共振频率也会略有不同。化学位移定义为同一种核在不同化学环境下共振频率的位移。

化学位移产生的原因是，分子中运动的电子受到外磁场的作用，产生感生电流。这一感生电流在核上产生感生磁场，感生磁场与外磁场相互叠加，使核上受到的有效场发生变化。这一现象称作核外电子的屏蔽作用。核外电子云密度越大，屏蔽效应越强，要发生共振吸收就要增加外磁场的强度，共振信号将移向高场区，反之将移向低场区。

### 脉冲付里叶变换

与红外、紫外及其它光谱技术相比，核磁共振的主要局限性是其固有的低灵敏度，这是由于在核磁共振实验中只涉及到很少量的核数目的变化造成的。对于除质子外的天然丰度低的核，灵敏度极低，用连续波核磁共振谱仪需要用累加器将波谱储存然后累加，这种累加将使信噪比增加n 倍，但是这样做扫描一张波谱所需要的时间较长，在这段较长的时间内，由于仪器的磁场漂移，将会对实验结果造成很大影响。于是后面出现了脉冲傅立叶变换核磁共振试验方法。

傅立叶变换核磁共振谱仪不是通过扫场或扫频产生共振，而是用恒定磁场激发，施加全频脉冲，产生共振，采集产生的感应电流信号，不同于连续播激发能够直接得到频谱，脉冲激发后得到的是不同共振频率自由感应衰减的相干图，是时域信号，需要经过傅立叶变换，获得一般核磁共振谱图。由于脉冲包含多个频率成分，因而射频脉冲激发相当于多频率激发，它可以使一个波谱范围内所有不同共振频率的核都同时得到激发，这就使波谱累加消耗时间的问题得到解决。

### 蛋白质结构测定

本次实验主要涉及通过$_{1}H－_{1}H$欧沃豪斯效应的顺序识别。
顺序识别是在已知蛋白质一级结构的基础上进行的。 具体做法是沿着多肽链的主链，交替使用***COSY***谱中残基内的$NH_{i} -H_{i}$耦合交叉峰及***NOESY*** 谱中第i个残基的$CH_{i}$和第i+1个残基的$NH_{i+1}$之间的***NOE*** 交叉峰。这可根据***COSY***谱和***NOESY***谱固有的对称性，将一个***NOESY***谱的左上三角形和一个***COSY***谱的右下三角形拼在一起，分析这个***COSY-NOESY***连接图。我们可以从经过残基自旋系统识别已确定的第i个残基在***COSY***谱上的$NH{i}$ 交叉峰出发，作垂直线和水平线，它们和图中对角线的交点分别是该残基的及$NH_{i}$质子的化学位移，然后再以质子化学位移为出发点作水平线，在***NOESY***谱中寻找$NH_{i}$和$NH_{i+1}$之间的***NOE***交叉峰。一旦找着了这个交叉峰，通过该峰作垂
线，该垂线与对角线的交点就是$NH_{i+1}$的化学位移，从这个$NH_{i+1}$的对角峰出发，在***COSY***谱上作水平线即可找到$NH_{i+1}$和$CαH_{i+1}$之间$J$耦合的交叉峰，如此继续下去，只要这个蛋白质的一级序列清楚，就可以完成整个多肽链质子的顺序识别。

- 数据处理

- 运行 Bruker，对部分读取的数据进行修正。对各维度的中心频率进行修正，比如，实验中第二维 N 的中心频率理论上为 H 的 1/10 左右，但错误读取成了 C 的中心频率，需要手动做修正为 60.818MHz。

- 接下来运行脚本文件，对 Bruker 的输出 FID 文件进行处理，从直接维开始做变换。

- 对 log 文件进行修改，将直接维和间接维的一阶和零阶相位改到实验对应的值，输入合适的输出文件路径和文件名。运行脚本，得到二维谱。运行 NMRDraw，右键左上角 FILE，打开刚生成的文件，得到二维谱图。选择右下角的谱峰，可以分别看水平和垂直的信号强度。可以看到，峰两侧的基线是不对称的，我们进行调相是基线两侧对称。调相位时，选择二维面上接近对角位置的两个峰，比如，实验中先调节左下角位置的峰的形状，改变 P0 使峰对称，按左键出现箭头将箭头对准峰尖位置，保证此峰相位不再改变；然后选择右上部分的与其他峰较为分离的峰，可以看出这个峰的基线仍然是不对称的，继续调节一节相位 P1，使基线对称。调节结束后，将得到的 P0、P1 的数值加到脚本文件的直接为相位 P0、P1 上，再次运行脚本，对输出文件重复上述操作，多次重复后，直到二维谱直接维任意谱峰位置基线对称。直接维调节好后，对间接维重复以上操作。

- 三维谱<br>
    运行 Sparky，打开 hsqc.ucsf、cbcaconh.ucsf、cbcanh.ucsf。首先，运行 ct 命令，对谱图的显示做调整，调节正峰负峰的颜色、阈值等。按 F8 框住需要区域进行选峰。对于三维图，不能直接在一张图上框选区域按 F8 选峰，如果只在一张图上操作则得到的只是这一张图上（第三维为某特定值）的峰，此时，需运用 RT 命令（region，定义空间）定义一个三维空间，选择 Z 的范围，则可在选中空间全部范围内积分选峰。对于 cbcaconh、cbcanh 两张谱图，两张图 H 一维的中心点可能选的不一样，运行 ST 命令，调其中一张谱的 H 的基准化学位移，使得两张谱图谱峰对齐。运行 YT 命令，使 cbcaconh、cbcanh 第三维（N）同步改变。
