# 1

## 实验目的

- 了解汽相扩散悬滴法的原理
- 熟练汽相扩散悬滴法操作

## 实验原理

蛋白质分子是一个复杂的体系，其构象会随外界条件的变化而变化。因此，虽然一般认为蛋白质分子与一般小分子一样，其结晶遵守一般的热力学规律，但是，小分子结晶的方法和技术却对蛋白质分子结晶不适用。目前蛋白质的结晶工作主要靠经验。蛋白质结晶的方法一般为：清澈的蛋白质溶液饱和溶液发生沉淀蛋白质的无定形沉淀。当满足适当条件，蛋白质就有可能以晶体形式从溶液中析出。

实验采用汽相扩散法进行晶体培养，在起始状态时，结晶母液与外液之间存在某些因素的梯度差，随着汽相扩散的进行，这些梯度差发生变化。如果变化后的条件合适，母液中的蛋白质就有可能以晶体状态析出。

- 准备母液及外液。其中母液中含有待结晶的蛋白及沉淀剂。外液中通常含有与母液中种类相同的沉淀剂，
- 使制备好的母液及外液共处一密封的小室，并且不使母液与外液直接接触。发生汽相扩散过程， 使得母液、外液中的总摩尔浓度趋于一致。
- 随着汽相扩散的进行，母液中的蛋白浓度会趋向过饱和状态，引起蛋白分子之间的聚集。若这种扩散体系的组成特性能使蛋白分子之间有序聚集－最初形成的这种微小聚集体－晶核就会在母液中形成。否则，母液中会出现蛋白的沉淀。 最后，随着扩散过程的进一步延续，这种合适的体系能使一些晶核生长发育成可适用的单晶。

# 2

## 实验目的

- 熟悉微晶接种法的操作；掌握蛋白质晶体收获的实验操作 。
- 熟悉晶体X-射线晶体衍射数据收集方法（in home）；了解X-射线单晶衍射仪的结构组成

## 实验原理

- 接种及晶体收获

    在很多情况下，直接从母液中获得的晶体,或者由于其尺寸大小的限制，或者由于其发育环境而导致的孪晶性，并不适用于X-射线衍射实验。一种有效的途径是进行结晶后改善（post-crystallization modification）。微晶接种法在一些时候对于晶体的改善非常有效。微晶接种法的基本原理是从已有的晶体制备微小的晶体或者晶核，有效的控制晶核的数量，然后在合适的沉淀剂条件下（一般是用略低于原先使用的沉淀剂的浓度），让晶核发育成合适的晶体。
    
    蛋白质晶体的收获是指将蛋白质晶体从结晶母液转移到特制的毛细石英管中（适用于常温下的衍射实验）或者转移到特制的loop装置而保存在液氮中（适用于冷冻条件下的衍射实验）的过程。在晶体收获时，要尽量避免在操作中由于机械力对蛋白质晶体造成损伤外；由于蛋白质晶体有大量水份构成。另外，现在越来越多的x-射线晶体衍射实验是在晶体冷冻条件下进行的，本实验只涉及冷冻（～100K）衍射实验条件下所要求的晶体收获，即晶体在收获需要冷冻保护液。
    
    冷冻保护液的选取一般是在原有的结晶条件的基础上添加适量的抗冻剂。多种用于结晶用的沉淀剂在一定范围内也可以直接用作抗冻剂，如20-30%PEG，30％以上的MPD，高浓度的NaCl， Li2SO4，蔗糖等。这样，如果晶体是从具有抗冻特性的液体中培养出来的，则可以将晶体从母液中取出后直接保存到液氮中。很多时候，会在原有的结晶条件的基将础上添加10－25％甘油，20—30％乙二醇作为抗冻剂。不同晶体所适合的抗冻剂也是不同的。


- 衍射数据收集

    衍射本质上是由于原子中的电子与x-射线的相互作用。X-射线（波长10-0.01nm）又称X-光，本质上是一种电磁波，当入射X-射线照射到晶体上时，晶体里原子中的电子会向空间中散射出与入射x-射线同频率的电磁波。由于这些（入射及出射）电磁波的干涉效应，使得空间中的某些位置（方向）因为波的相互叠加而观测到到衍射点。因此，晶体衍射理论中会出现晶体中电子散射体系，原子散射因子，晶胞散射因子（结构因子）及整个晶体的衍射。蛋白质晶体学的主要任务是得到一个晶胞中的电子密度在空间上的分布，从而得到蛋白质的结构。
    
    X-射线的来源主要有两种：一种是在室内X射线仪上使用，通过高能电子流轰击铜靶（或铬靶、钼靶等），产生多个特征波长的X射线，常用的铜靶CuKα的波长为1.5418Å；另一种就是利用同步辐射所产生的X射线，其波长可以变化。

## 实验原理
衍射数据处理包括：指标化，强度积分、合并、振幅的还原。
指标化的基本步骤：
- 挑选若干个强衍射斑点并测定其机械坐标
    - 强弱的选取标准
    - 衍射分辨率的上下限
- 计算出若干个可能的单位晶胞
	判断哪个是正确的解
- 调整修正各种参数
    - 晶体的镶嵌度因子
    - 衍射斑点强度积分及其误差的周边范围大小
    - 晶体至探测器平面的距离
    - 其它需要修正的机械参数

- 简要说明正确空间群的理由

    对于正确的空间群，运行molrep所得的TF/σ、score、contrast等数值较大。
    在实验part1中，用P43212空间群所得结果明显优于用P41212空间群，正确的解为P43212；
    part2中，将空间群P6522改为P6122后，所得结果明显变差，正确的解为P6122


- 对Log文件（含求解相位及修正时）输出参数做简单总结。

    TF/σ、score、contrast、corrF、FinalCC等数值越大，说明空间群符合越好。

## 修正过程
- 启动coot，打开结构文件lyso_MR+inh.pdb及密度图lyso_refmac1.mtz。
- 通过coot菜单栏Validate->Density fit analysis 查看结构与电子密度整体符合情况。
- 按主链顺序依次调整结构与电子密度符合不好的残基主链与侧链构象
- 添加和修正抑制剂小分子结构。
- 输出第一轮修正好的结构文件：lyso_coot-#.pdb
- 使用ccp4中的refmac5进行结构位置修正：
- 输入lyso_coot-#.pdb及lyso_refmac1.mtz文件，设定分辨率、修正轮次数Do _ cycles等参数。
- 重复1-6步反复迭代修正结构。在R Rfree 因子下降到合适值时尝试添加水分子coot->Find Waters
- 重复1-6，反复迭代修正，删除多余水分子，直到R factor，Rfree降到目标值。
