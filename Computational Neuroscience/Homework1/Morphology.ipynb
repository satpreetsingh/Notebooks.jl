```julia
data = readdlm("j8_L23pc.CNG.swc.txt");
# data = readdlm("Purkinje-slice-ageP43-6.CNG.swc.txt");

# branching point
ids,x,y,z,fids = Int.(data[:,1]),data[:,3],data[:,4],data[:,5],Int.(data[:,7]);
degree = zeros(ids)
for (id,fid) in zip(ids,fids) # fids => father ids
    fid>0 && (degree[fid] += 1)
end
@show sum(degree.>=2)

# Plotting
using Plots; pyplot()

p = scatter(x,y,z,markersize=0.5,legend=false)
for (id,fid) in zip(ids,fids)
    fid>0 && plot!(p,[x[fid],x[id]],[y[fid],y[id]],[z[fid],z[id]],c=:blue)
end
p
```
