### Problem:  $\psi (x)=\sum _{k=1}^{\infty } \frac{1}{k (k+x)}$

Analytically, we could get $\psi (x)=\sum _{k=1}^{\infty } \frac{1}{k (k+x)}$, $\psi (1)=\sum _{k=1}^{\infty } \frac{1}{k (k+1)}=1$

#### Mothod 1: Direct Mothod

```julia
a(k,x)=1./(k.*(k.+x))
ψ(N,x)=sum(a(1:N,x));
```

```julia
x=[0:0.1:1;]
plot(x,map(x->ψ(100,x),x))
plot(x,map(x->ψ(1000,x),x))
plot(x,map(x->ψ(10000,x),x))
legend(["100","1000","10000"]);
```

```julia
for x in [0.0,0.1,0.5,1.0,√3,10,100,200]
    answer=ψ(10000000,x)
    @printf "x=%s  ψ(x)=%6.2f,%16.12e\n" string(x) answer answer
end
```

#### Mothod 2: Difference Mothod

$\psi (0)-\psi (x)=\sum _{k=1}^{\infty } \left(\frac{1}{k^2}-\frac{1}{k (k+x)}\right)=\sum _{k=1}^{\infty } \frac{x}{k^2 (k+x)})$, so it will converge faster

```julia
a(k,x)=x./(k.^2.*(k.+x))
ψ(N,x)=π^2/6-sum(a(1:N,x));
```

```julia
x=[0:0.1:1;]
plot(x,map(x->ψ(100,x),x))
plot(x,map(x->ψ(1000,x),x))
plot(x,map(x->ψ(10000,x),x))
legend(["100","1000","10000"]);
```

```julia
for x in [0.0,0.1,0.5,1.0,√3,10,100,200]
    answer=ψ(10000000,x)
    @printf "x=%s  ψ(x)=%6.2f,%16.12e\n" string(x) answer answer
end
```

### Conclusion
The second method will converge faster than the first one.
